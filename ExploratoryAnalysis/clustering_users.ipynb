```python
def create_hashtable(df):
    """
    Creates a hashtable that contains all the clusters as keys and all the users that are within the cluster as values in a list format.

    Args: df: dataframe with all the users and labeled with the cluster they have been organized to

    returns : dictionary {(cluster number) : ([user 1, ..., user n])}

    """

    clusters = df.groupby("cluster")["user_id"].apply(list)
    
    hashtable = {cluster: user for cluster,user in zip(clusters.index, clusters.values)}
    return hashtable
```
